```python
# ============================================
# 1. Import Required Libraries
# ============================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Settings for better visuals
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,6)

# ============================================
```

```python
from google.colab import files

# Upload dataset
data_file = files.upload()
df = pd.read_csv(next(iter(data_file)))
```

```python
# 3. Dataset Overview
# ============================================
print("First 5 rows of the dataset:")
display(df.head())

print("\nDataset Information:")
df.info()

print("\nStatistical Summary:")
display(df.describe())
```

```python
# ============================================
# 4. Check for Missing Values
# ============================================
print("\nMissing Values in Each Column:")
print(df.isnull().sum())
```

```python
# ============================================
# 5. Check for Duplicates
# ============================================
print("\nNumber of duplicate rows:", df.duplicated().sum())

# ============================================
# 6. Data Types
# ============================================
print("\nData types of columns:")
print(df.dtypes)
```

```python
# ============================================
# 7. Univariate Analysis
# ============================================

# Categorical Variables
categorical_cols = ['gender', 'race_ethnicity', 'parental_level_of_education', 'lunch', 'test_preparation_course']

for col in categorical_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index, palette='Set2')
    plt.title(f"Distribution of {col.replace('_', ' ').title()}")
    plt.xticks(rotation=45)
    plt.show()

# Numerical Variables
numerical_cols = ['math_score', 'reading_score', 'writing_score']

for col in numerical_cols:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f"Distribution of {col.replace('_', ' ').title()}")
    plt.xlabel(col.replace('_', ' ').title())
    plt.ylabel('Frequency')
    plt.show()
```

```python

# ============================================
# 8. Bivariate Analysis
# ============================================

# Gender vs Average Scores
df['average_score'] = df[['math_score', 'reading_score', 'writing_score']].mean(axis=1)

plt.figure(figsize=(8,5))
sns.boxplot(x='gender', y='average_score', data=df, palette='Pastel1')
plt.title('Average Score Distribution by Gender')
plt.show()

# Lunch vs Average Scores
plt.figure(figsize=(8,5))
sns.boxplot(x='lunch', y='average_score', data=df, palette='Pastel2')
plt.title('Average Score Distribution by Lunch Type')
plt.show()

# Test Preparation vs Scores
plt.figure(figsize=(8,5))
sns.boxplot(x='test_preparation_course', y='average_score', data=df, palette='Set3')
plt.title('Average Score by Test Preparation Course')
plt.show()
```

```python
# ============================================
# 9. Correlation Analysis
# ============================================
plt.figure(figsize=(8,6))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Scores')
plt.show()
```

```python
# ============================================
# 10. Key Insights
# ============================================
print("""
🔹 Students who completed test preparation scored higher on average.
🔹 Standard lunch students generally perform better than free/reduced lunch students.
🔹 Math, reading, and writing scores are highly positively correlated.
🔹 Gender differences exist: females tend to score higher in reading and writing.
""")
```
